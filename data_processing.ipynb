```python
import pandas as pd
from tqdm import tqdm_notebook
from eunjeon import Mecab
from konlpy.tag import Okt
mecab = Mecab()
okt = Okt()
```

# 에러 없애기

```python
df = pd.read_csv('../data/자소서_잡코추가.csv',index_col=0, engine='python', encoding='utf8')
```

```python
df = df.reset_index().drop(columns=['index'])
```

```python
df.head()
```

```python
debug = []
bugidx = []
i = 0
for txt in tqdm_notebook(df.jss):
    nouns = okt.nouns(txt)
    if len(nouns) == 0:
        print(i, txt)
        bugidx.append(i)
    debug.append(nouns)
    i += 1
```

```python
len(df)
```

```python
df = df.drop(bugidx)
```

```python
df.to_csv('../data/자소서_잡코추가2.csv')
```

```python
cadf.head()
```

```python
df = pd.read_csv('../data/자소서5.csv',index_col=0, engine='python', encoding='utf8')
```

```python
df = df.merge(cadf,left_on='category', right_on='category2')[['category1','category2','jss']]
```

```python
df = df[df.category2 != '정보통신·IT']
```

```python
df.groupby('category1').agg('count')
```

```python
labels = []
for i in df.category1.tolist():
    if i == '경영기획':
        labels.append(0)
    elif i == '금융':
        labels.append(1)
    elif i == '기술영업':
        labels.append(2)
    elif i == '마케팅':
        labels.append(3)
    elif i == '물류':
        labels.append(4)
    elif i == '사무':
        labels.append(5)
    elif i == '생산품질':
        labels.append(6)
    elif i == '영업':
        labels.append(7)
    elif i == '인사':
        labels.append(8)
    elif i == '재무':
        labels.append(9)
    elif i == '통계':
        labels.append(10)
    elif i == '응용프로그래밍':
        labels.append(11)
    elif i == '서버·시스템':
        labels.append(12)
```

```python
df['label'] = labels
```
